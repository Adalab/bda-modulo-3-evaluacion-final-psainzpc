```python
# después de mirar las medias de los clientes con niveles de estudios superior (bachelor) e inferior (high school or below)
# hago una copia del df para probar a impotar por diferentes métodos y así elegir cual tendría más logica imputar
```

```python
df_loyalty_copia = df_loyalty[df_loyalty["Education"].isin(["College", "High School or Below", "Bachelor"])]
```

```python
# hemos dicho que para las tres columnas restantes aplicaremos los dos métodos para luego compararlos, empezaremos por el IterativeImputer
# instanciamos las clases
imputer_iterative = IterativeImputer(max_iter = 20, random_state = 42)

# ajustamos y tranformamos los datos
imputer_iterative_imputado = imputer_iterative.fit_transform(df_loyalty_copia[["Salary"]])

# comprobamos que es lo que nos devuelve, que en este caso es un array también
imputer_iterative_imputado
```

```python
df_loyalty_copia["Salary_iterative"] = imputer_iterative_imputado

print(f"Después del 'Iterative' tenemos: \n{df_loyalty_copia["Salary_iterative"].isnull().sum()} nulos")
```

```python
df_loyalty_copia[df_loyalty_copia["Education"] == "College"]["Salary_iterative"].mean()
```

```python
# hagamos ahora lo mismo con el KNNImputer. NOTA, la ejecución de esta celda puede tardar un ratito
#salary_column = df_loyalty_copia[['Salary']]
# instanciamos la clase del KNNImputer
imputer_knn = KNNImputer(n_neighbors = 5)

# ajustamos y transformamos los datos
imputer_knn_imputado = imputer_knn.fit_transform(df_loyalty_copia[['Salary']])

# comprobamos que es lo que nos devuelve, que sigue siendo un array
imputer_knn_imputado
```

```python
# por último nos queda añadir ese array al DataFrame como hemos hecho hasta ahora
df_loyalty_copia["Salary_knn"] = imputer_knn_imputado


# comprobamos los nulos
print(f"Después del 'KNN' tenemos: \n{df_loyalty_copia["Salary_knn"].isnull().sum()} nulos")
```

```python

# Filtrar los salarios de los grupos "High School or Below" y "Bachelor"
salarios_sup_inf = df_loyalty[df_loyalty["Education"].isin(["High School or Below", "Bachelor"])]["Salary"]

# Calcular el salario promedio de esos grupos
salario_promedio = salarios_sup_inf.mean()

# Imputar el salario promedio en los registros con nivel de estudios "College" y salario nulo
df_loyalty["Salary"] = df_loyalty["Salary"].fillna(salario_promedio)
```
